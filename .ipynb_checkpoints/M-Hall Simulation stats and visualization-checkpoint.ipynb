# Monty Hall Problem
### The following stats is to show whether the statistics behind the monty hall problem is true 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

### Data Preparation

```python
# reading in data as a json file
raw_data = pd.read_json('data.json')
raw_data_strategy_1 = raw_data['strategy_1']
raw_data_strategy_2 = raw_data['strategy_2']

stra_1_series = pd.Series(raw_data_strategy_1['outcome'], index=raw_data_strategy_1['game_no'], name='stra_1_outcome')
#stra_1_series

stra_2_series = pd.Series(raw_data_strategy_2['outcome'], index=raw_data_strategy_2['game_no'], name='stra_2_outcome')

df =  pd.concat([stra_1_series, stra_2_series ], axis=1)
df.head(10)
```

## Statistics on Data From Both Strategies

```python
df.describe()
```

```python
df.corr()
```

## Statistics on Strategy 1, when user maintains first choice

```python
stra_1_lost_outcomes = (stra_1_series==0).sum()
stra_1_won_outcomes = (stra_1_series==1).sum()

print('Stra_1_lost outcomes: ', stra_1_lost_outcomes)
print('Stra_1_won outcomes: ', stra_1_won_outcomes)
```

```python
# plotting number of losses and wins on a histogram
y_vals = [stra_1_lost_outcomes, stra_1_won_outcomes]
x_vals = [0,1]

plt.subplot()
plt.bar(x_vals, y_vals, color='rb')

plt.xticks(x_vals, ['lost', 'won'], fontsize=20)

plt.ylabel('number of outcomes')
plt.xlabel('Game outcomes, 0 or 1')
plt.title('Chart of the wins and losses in Strategy one', fontsize= 25 )
plt.show()
```

## Statistics on Strategy 2, when user changes his first choice

```python
stra_2_lost_outcomes = (stra_2_series==0).sum()
stra_2_won_outcomes = (stra_2_series==1).sum()

print('Strategy_2_lost outcomes: ', stra_2_lost_outcomes)
print('Strategy_2_won outcomes: ', stra_2_won_outcomes)


#plotting number of losses and wins on a histogram
y_vals = [stra_2_lost_outcomes, stra_2_won_outcomes]
x_vals = [0,1]

plt.subplot()
plt.bar(x_vals, y_vals, color='rb')

plt.xticks(x_vals, ['lost', 'won'], fontsize=20)

plt.ylabel('number of outcomes')
plt.xlabel('Game outcomes, 0 or 1')
plt.title('Bar Chart of the wins and losses in Strategy Two', fontsize= 25)
plt.show()
```

# Plotting charts of Both Strategies on the same plot

```python
plt.figure(2)

ax1 = plt.subplot(211)

stra_1_y_vals = [stra_1_lost_outcomes, stra_1_won_outcomes]
stra_1_x_vals = [0,1]

plt.bar(stra_1_x_vals, stra_1_y_vals, color='rb')
plt.xticks(x_vals, ['lost', 'won'], fontsize=20)

plt.ylabel('number of outcomes')
plt.xlabel('Game outcomes, 0 or 1')
plt.title('Strategy one', fontsize= 25 )
plt.show()

plt.subplot(212)
stra_2_y_vals = [stra_2_lost_outcomes, stra_2_won_outcomes]
stra_2_x_vals = [0,1]

plt.bar(stra_2_x_vals, stra_2_y_vals, color='rb')
plt.xticks(x_vals, ['lost', 'won'], fontsize=20)

plt.ylabel('number of outcomes')
plt.xlabel('Game outcomes, 0 or 1')
plt.title('Strategy Two', fontsize= 25 )

plt.show()
```

#  Conclusion
### Proving the Stats

```python
#Probability of win with Strategy One
prob_of_win_stra_1 = stra_1_won_outcomes/(stra_1_won_outcomes + stra_1_lost_outcomes)


#probability of win with Strategy Two
prob_of_win_stra_2 = stra_2_won_outcomes/(stra_2_won_outcomes + stra_2_lost_outcomes)

print(" Probability of Winning win with Strategy One is : ", prob_of_win_stra_1)

print(" Probability of Winning win with Strategy Two is : ", prob_of_win_stra_2)
```
